# 2. How to deal with missing values?

When we inspect a dataset, it is a good idea to check if we have missing values (NaN) because: 
- especially for Machine Learning, missing values often are a problem, so we have to fill them or drop the samples. 
- Missing values can also tell us something themselves, e.g. if many people didn't give an answer to a certain question, that is information in itself.

```python
# To check for missing values, we can use the .isna()-function:
complete_df.isna()
```

```python
# Here you see False/True values that indicate if there is a missing value (True). 
# But this is not very readable for us. 
# So a better way to inspect the missing values is plotting this dataframe as a heatmap: 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
sns.heatmap(complete_df.isna(), cbar=False)
```

```python
# Here we can now see, that most of the missing values (where it is light) are in the "comments"-column. 
# There also seem to be a couple of penguins with missing data in more columns (the white lines)
```

```python
# So in this case, lets drop the Comments-column, because there is only a little data in there:
complete_df.drop('Comments', axis=1, inplace=True)
```

```python
# and check that the column is deleted:
complete_df.head()
```

```python
# Lets check again what our heatmap looks like now: 
```

```python
sns.heatmap(complete_df.isna(), cbar=False)
```

```python
# Now we can decide to also drop the two penguins that have missing data: 
complete_df.dropna(inplace=True)   
#By default, this method drops all rows that have some missing data. 
# But you can adjust this with parameters (check the documentation), e.g. "thresh"
```

```python
# Lets check again: 
sns.heatmap(complete_df.isna(), cbar=False)
```

```python
# A nice black square :-) 
```

## Just deleting the samples and columns that have missing values is not always the best choice!
What we can also do, is fill the missing values with customized values.   
- E.g. we can fill the Nans with zeros: `complete_df.fillna(0)`   
- Or we can fill the NaNs in a specific column with the mean of the column:   
`complete_df['Flipper Length (mm)'].fillna(complete_df['Flipper Length (mm)'].mean(), inplace=True)`  

There are many different options to fill NaNs - we will revisit this topic in the next weeks!
